```python
# 問16. for文

# 16-1
genes = ["AT1G01010", "AT1G01020", "AT1G01030"]
for g in genes:
    print(g)

# 16-2
for i in range(1, 4):
    print(f"rep{i}")

# 16-3
genes = ["AT1G01010", "AT1G01020"]
treatments = ["Mock", "UV"]

for g in genes:
    for t in treatments:
        print(f"{g} - {t}")

# 16-4 辞書と JA-responsive
pathways = {
    "AT1G01010": "SA pathway",
    "AT1G01020": "JA pathway",
    "AT5G05010": "SA pathway",
    "AT3G01110": "SA pathway",
}

for gene_id, pathway in pathways.items():
    print(f"{gene_id} -> {pathway}")
    if pathway == "JA pathway":
        print("JA-responsive")
```

```python
# 問17 & 18. while & break

# 17-1 カウントダウン
n = 5
while n > 0:
    print(n)
    n -= 1

# 18-1 10周目でループを抜ける
count = 1
while True:
    print(f"今は {count} 周目です")
    if count == 10:
        break
    count += 1

# 18-2 見つかったところで break
genes = ["AT1G01010", "AT1G01020", "AT1G01030"]

for g in genes:
    if g == "AT1G01020":
        print("見つかりました")
        break
    
```

```python
# 問19. continue と break

# 19-1 奇数だけ表示
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i)

# 19-2 正の数だけ合計
values = [1, -2, 3, -4, 5]
total = 0
for v in values:
    if v < 0:
        continue
    total += v
print("正の数の合計:", total)

# 19-3 空文字をスキップ
genes = ["AT1G01010", "", "AT1G01020"]
for g in genes:
    if g == "":
        continue
    print(g)

# 19-4 5 で停止
for i in range(1, 11):
    if i == 5:
        print("5 で停止")
        break
    print(i)
```

```python
# 問22. 例外処理（try / except）

def function(a, b):
    try:
        result = a / b
        print(result)
        return result
    except ZeroDivisionError:
        print("0 で割ることはできません")
        print(None)
        return None

# 動作確認
function(10, 2)
function(10, 0)
```
